```python
import pandas as pd
import pyramid
import plotly.plotly as plt
import cufflinks as cf
import numpy as np
from plotly.plotly import plot_mpl
from statsmodels.tsa.seasonal import seasonal_decompose
from pandas import *
from pandas import ExcelWriter
from pandas import ExcelFile
from pandas import datetime
from matplotlib import pyplot
```

```python
df=pd.read_excel('DC Shipment 1 Cleaned (1).xlsx',sheetname='DC Shipments Monthly Data Clean')
```

```python
df.head()
```

```python
df1=df.groupby(['NEW_NAME_MERGE','delivery_date']).agg({'sum_quantity_adj_edit':'sum'})
#len(df1)
newlist=[]
list_1=list(df.NEW_NAME_MERGE)
list_1
for i in list_1:
    if i not in newlist:
        newlist.append(i)
newlist
df1.head()
```

```python
m1='2017-03-31'
m2='2017-04-30'
m3='2017-05-31'
for i in range(len(newlist)):
    if m2 in df1.loc[newlist[i]].index:
        if m3 in df1.loc[newlist[i]].index:
            df1.loc[newlist[i]].loc[m2]=int(1/2*(df1.loc[newlist[i]].loc[m1]+df1.loc[newlist[i]].loc[m3]))
        else:
            df1.loc[newlist[i]].loc[m2]=df1.loc[newlist[i]].loc[m1]
```

```python
df1.loc[newlist[1]]
```

```python
df1['NEW_NAME_MERGE']=df1.index.get_level_values('NEW_NAME_MERGE')
df1['delivery_date']=df1.index.get_level_values('delivery_date')
cols = ['NEW_NAME_MERGE','delivery_date','sum_quantity_adj_edit']
df1=df1[cols]
df2=df1.reset_index(drop=True)
```

```python
writer = pd.ExcelWriter('newdf.xlsx', engine='xlsxwriter')

# Convert the dataframe to an XlsxWriter Excel object.
df2.to_excel(writer, sheet_name='Sheet1')

# Close the Pandas Excel writer and output the Excel file.
writer.save()
```
